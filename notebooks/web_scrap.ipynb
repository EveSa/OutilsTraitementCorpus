```python
def get_first_header_text(root):
    text = []
    for paragraph in root.xpath("//div[@id='bodyContent']//p|//div[@id='bodyContent']//h2"):
        
        if paragraph.tag != "h2":
            text.append(" ".join(paragraph.xpath(".//text()")))
        else:
            return " ".join(text).replace("\n", "")

get_first_header_text(root)
```
